```python
import pandas as pd
import matplotlib.pyplot as plt

labels = []
shows = []
for i in range(30):
#for name in test['Name']:
    labels.append(df[test['Name'][i]]['sarcasm'])
    shows.append(df[test['Name'][i]]['show'])
            
            
# Sample dataframe
data_filtered = {
    'names': shows,
    'labels': labels
}
df_show = pd.DataFrame(data_filtered)

# Group by names and labels to get the count of each combination
grouped = df_show.groupby(['names', 'labels']).size().reset_index(name='counts')

colors = ['#F9DBBD', '#F9C784', '#FF8C42', '#B88C9E', '#E2E4F6', '#E2E4F6', '#60B2E5']
# Plot pie chart
fig, ax = plt.subplots(figsize=(20, 15))
ax.pie(grouped['counts'], labels=[f"{row['names']} - {row['labels']}" for _, row in grouped.iterrows()],
        autopct='%1.1f%%', startangle=90, colors=colors)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

#plt.savefig('tv_eda.png', dpi=500)
#plt.title("Percentage of each name + label")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

labels = []
shows = []
utterances = []
for name in df.keys():
    labels.append(df[name]['sarcasm'])
    shows.append(df[name]['show'])
    utterances.append(df[name]['utterance'])
            
            
# Sample dataframe
data_filtered = {
    'names': shows,
    'labels': labels,
    'utterances': utterances
}
df_utterance = pd.DataFrame(data_filtered)

df_utterance['word_count'] = df_utterance['utterances'].apply(lambda x: len(x.split()))

fig, ax = plt.subplots(sharex=True, figsize=(20, 15))
sns.histplot(data=df_utterance, x='word_count', hue='labels', 
             multiple='stack', bins=range(1, df_utterance['word_count'].max() + 2), 
             discrete=True, palette = ['#4357AD', '#E6DB7E'])
plt.grid(axis='y')
plt.xlabel('Word Count', fontsize=40)
plt.ylabel('Counts', fontsize=40)
plt.tick_params(axis='x', which='major', labelsize=30)
plt.tick_params(axis='y', which='major', labelsize=30)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.legend(['Sarcastic', 'Non-Sarcastic'], frameon=False, fontsize=35)

plt.savefig('tv_eda_utterance_human.png', dpi=500)
# #plt.title("Percentage of each name + label")
# plt.show()
```

```python
labels = []
characters = []
for name in df.keys():
    if df[name]['show'] == 'FRIENDS':
        labels.append(df[name]['sarcasm'])
        #characters.append(df[name]['speaker'])
        
        if df[name]['speaker']== 'PERSON' or df[name]['speaker'] == 'PERSON1':
            characters.append('OTHERS')
        else:
            characters.append(df[name]['speaker'])
label_list = [int(val) for val in labels]
df_new = pd.DataFrame(
    {'names': characters,
     'labels': label_list
    })

df_grouped = df_new.groupby(['names', 'labels']).size().reset_index(name='counts')

sort_key = df_grouped[df_grouped['labels'] == 1].set_index('names')['counts'].to_dict()

# Sort df_grouped using the sort_key
df_grouped['sort_order'] = df_grouped['names'].map(sort_key)
df_grouped = df_grouped.sort_values(by='sort_order', ascending=False).drop(columns='sort_order')

# Plot the sorted df_grouped
fig, ax = plt.subplots(sharex=True, figsize=(28, 20))
sns.barplot(x='names', y='counts', hue='labels', data=df_grouped, hue_order = [1, 0], palette = ['#E6DB7E', '#4357AD'])
plt.tick_params(axis='x', which='major', labelsize=30, rotation=90)
plt.tick_params(axis='y', which='major', labelsize=45)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.legend(fontsize=50, frameon=False)

plt.savefig('eda_1.png', dpi=500)
plt.show()
```
