```python
import json
import seaborn as sns
import matplotlib.pyplot as plt
with open('sarcasm_data.json', 'r') as file:
    data = json.load(file)
```

```python
import pandas as pd
test = pd.read_csv('human_test.csv')
```

```python
df = pd.DataFrame(data)
```

```python
labels = []
characters = []
for name in df.keys():
   # if df[name]['speaker'] == 'PENNY':
    labels.append(df[name]['sarcasm'])
    characters.append(df[name]['speaker'])
label_list = [int(val) for val in labels]
df_new = pd.DataFrame(
    {'names': characters,
     'labels': label_list
        
    })
```

```python
df_grouped = df_new.groupby(['names', 'labels']).size().reset_index(name='counts')
fig, ax = plt.subplots(sharex=True, figsize=(28, 20))
sns.barplot(x='names', y='counts', hue='labels', data=df_grouped)
plt.tick_params(axis='x', which='major', labelsize=30, rotation=90)
plt.tick_params(axis='y', which='major', labelsize=30)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.legend(fontsize=50, frameon=False)
```
